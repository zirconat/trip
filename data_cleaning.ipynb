```python
import streamlit as st
import pandas as pd
import calendar
import plotly.express as px
import altair as alt

df = pd.read_excel("Trip and visit WY24_25 Database.xlsx")
df
```

```python
# add month & year column
df['Month'] = df["Departure Date"].dt.month_name()
df['Year'] = df["Departure Date"].dt.year
df
```

```python
# Group by Year, Month , Type, Del Lead
group_data1 = df.groupby(['Year','Month', 'Type', 'Del Lead']).size().reset_index(name='count')
group_data1
```

```python
group_data1.dtypes
```

```python
# Pivot the data for plotting
pivot_df = group_data1.pivot_table(index=['Year', 'Month', 'Type'], columns='Del Lead', values='count', fill_value=0)
pivot_df
```

```python
# Create a multi-index for plotting
pivot_df.index = pd.MultiIndex.from_tuples(pivot_df.index, names=['Year', 'Month', 'Type'])
pivot_df.index
```

```python
# using altair
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Month:O', title='Month'),
    y=alt.Y('count()', title='No. of Trips & Visits'),
    color='Del Lead',
    tooltip=['Month', 'Del Lead', 'Type', 'count()']
).facet(
    column='Year',
    title='Number of Events by Month and POC'
).interactive()

chart.show()
```

```python
# filter data for 'Chief'
chief_filtered = df[df['Del Lead'] == 'Chief']

# Group and cound by type
chief_df = chief_filtered.groupby(['Type']).size().reset_index(name='count')

# Calculate total count
chief_total = chief_df['count'].sum()

# add column for percentage calculation
chief_df['Percentage'] = chief_df['count'] / chief_total * 100

# Create donut chart
chief_chart = alt.Chart(chief_df).mark_arc(outerRadius=80, innerRadius=60).encode(
    theta = alt.Theta('Percentage:Q', stack = True),
    color = 'Type',
    tooltip = ['Type', 'count']
)
chief_chart.show()
#st.altair_chart(bb_chart)
```

```python
del_df = df.groupby(['Del Lead']).size().reset_index(name='count')
del_df
```

```python
del_total = del_df['count'].sum()
del_df['Percentage']=del_df['count'] / del_total * 100
del_df
```

```python
del_chart = alt.Chart(del_df).mark_arc(outerRadius=80, innerRadius=60).encode(
        theta = alt.Theta('Percentage:Q', stack = True),
        color = 'Del Lead',
        tooltip = ['Del Lead', 'count']
    )
del_chart.show()
```
